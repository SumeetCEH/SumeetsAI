```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 8, 12, 20, 26, 38])

# Provide 6 colors for 6 points
mycolor = ['red', 'green', 'purple', 'blue', 'orange', 'cyan']

plt.scatter(x, y, color=mycolor)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["A","B","C","D"])
y = np.array([0, 8, 12, 20])

plt.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([35,25,25,15])

plt.pie(x)
plt.show()
```

**labels**

```python
x = np.array([170,10,250,300])
mylabels = ["A","B","C","D"]
plt.pie(x,labels = mylabels)

plt.show()
```

```python
x = np.array([170,10,250,300])
mylabels = ["A","B","C","D"]
plt.pie(x,labels = mylabels,startangle=90)

plt.show()
```

```python
plt.pie(x,labels = mylabels,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
x = np.array([170,10,250,300])
mylabels = ["A","B","C","D"]
myexplode=[0.2,0,0,0]
plt.pie(x,labels = mylabels,startangle=90,explode=myexplode)

plt.show()
```
