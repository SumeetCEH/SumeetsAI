```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species']=le.fit_transform(df['species'])
```

```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
import keras
```

```python
model=keras.Sequential()
```

```python
model.summary()
```

```python
model.add(keras.Input(shape=(4,),batch_size=32,))
```

```python
model.add(keras.layers.Dense(
    32,
    activation='relu'
))
```

```python
model.add(keras.layers.Dense(
    3,
    activation='softmax'
))
```

```python
model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
callback=keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=3,
    verbose=0,
    mode="auto",
    restore_best_weights=True)
```

```python
model.fit(
    x=X_train,
    y=y_train,
    batch_size=32,
    epochs=1000,
    verbose="auto",
    validation_split=0.0,
    validation_data=(X_test,y_test),
    callbacks=[callback]
)
```

```python
model.evaluate(
    x=X_test,
    y=y_test,
    batch_size=32,
    verbose="auto"
)
```
