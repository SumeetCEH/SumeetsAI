```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 25, 20, 15]

# Create pie chart with start angle and shadow
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,     # Start from the top (default is 0°)
    shadow=True        # Adds shadow for 3D effect
)

plt.title('Programming Language Popularity')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 25, 20, 15]

# Explode the first slice (Python)
explode = [0.1, 0, 0, 0]  # 0.1 means "explode" the first slice

# Create pie chart
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    explode=explode
)

plt.title('Programming Language Popularity')
plt.axis('equal')  # Draw as a circle
plt.show()

```
